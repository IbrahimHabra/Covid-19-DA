```python
df4 = df3.copy()
```

```python
# print(df4.dtypes)
for i in df4.columns:
    if (df4.dtypes[i] == 'float'):
        df4[i].fillna(value=df4[i].mean(), inplace=True)
```

```python
df4.info()
```

```python
df5 = df4.drop('Province/State', axis=1)
```

```python
df5.info()
```

```python
df5.drop_duplicates(inplace=True)
df5['Deaths'] = df5['Deaths'].astype(np.int64)
df5['Recovered'] = df5['Recovered'].astype(np.int64)
df5['Confirmed'] = df5['Confirmed'].astype(np.int64)
df5.dtypes
```

```python
df5.head()
```

```python
df5n = df5.select_dtypes(np.number)
df5c = df5.select_dtypes('object')
```

```python
fig, axes = plt.subplots(len(df5n.columns), 2, figsize=(12, 5 * (len(df5n.columns))))

for i, col in enumerate(df5n.corr(numeric_only=True).columns):
    # Histogram with KDE


    df5n[col].plot(kind='hist', density=True, ax=axes[i, 0], bins=20, alpha=0.5, color='blue')
    df5n[col].plot(kind='kde', ax=axes[i, 0], color='red')
    axes[i, 0].set_title(f'{col} - Histogram with KDE')

    # Boxplot
    df5n.boxplot(column=col, ax=axes[i, 1])
    axes[i, 1].set_title(f'{col} - Boxplot')

plt.tight_layout()
plt.show()
```
